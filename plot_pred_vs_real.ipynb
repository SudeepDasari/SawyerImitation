# Description
This code will plot the real trajectories taken by the sawyer robot demonstration vs ones predicted by the CNN given video frames.

```python
#imports
import numpy as np
import cPickle as pickle
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pkl_name = 'real_vs_pred.pkl'
batches = pickle.load(open(pkl_name, 'rb'))
real_actions = batches['reals']
predicted_actions = batches['predictions']
```

```python
batch_to_plot = 6
N_JOINTS = 7
for i in range(N_JOINTS):
    plt.figure()
    plt.title("Real Actions (Red) vs Predicted Actions (Blue) for Joint: " +str(i))
    plt.plot(real_actions[batch_to_plot][:, i], color = 'r')
    plt.plot(predicted_actions[batch_to_plot][:, i], color = 'b')
    plt.savefig('joint'+str(i)+'plot.png')
```
